## Soal 1

Gunakan tabel order_items untuk mengetahui total dari beberapa status pengiriman!

```python
from sqlalchemy import create_engine
import pymysql
import pandas as pd

# membuat engine
engine = create_engine("mysql+pymysql://{user}:{pw}@localhost/{db}"
                       .format(user="bakti",
                               pw="123",
                               db="jne2"))

# engine koneksi ke database
db_enco = engine.connect()

# Perintah SQL 
sql = "SELECT status, COUNT(status) As Total FROM order_items GROUP BY status"

# Eksekusi perintah  SQL dengan Python (panda)
pd.read_sql(sql,db_enco)
```

## Soal 2

Pada tabel inventory_items. Tampilkan informasi kategori, nama produk dan biaya (dengan kondisi pembulatan dua angka dibelakang koma) tanpa duplikat dengan memfilter berdasarkan kategori produk (bebas pilih)

```python

# Perintah SQL 
sql = "SELECT DISTINCT product_category, product_name, ROUND(cost, 2) PembulatanCost \
FROM inventory_items \
WHERE product_category = 'Tops & Tees';"

       
# Eksekusi perintah  SQL dengan Python (panda)
pd.read_sql(sql,db_enco)
```

## Soal 3

Pada tabel products, urutkan dari yang terbesar rata-rata harga retail berdasarkan jenis brand      

```python
# Perintah SQL 
sql = "SELECT brand, AVG(retail_price) RataHarga \
FROM `products` \
GROUP BY brand \
ORDER BY RataHarga DESC;"

# Eksekusi perintah  SQL dengan Python (panda)
pd.read_sql(sql,db_enco)
```

## Soal 4

Kelompokkan inventory_items berdasarkan kategori produk lalu, filter dengan kondisi Total Biaya lebih dari 150
 

```python
# Perintah SQL 
sql = "SELECT product_category, SUM(cost) TotalCost \
FROM `inventory_items` \
GROUP BY product_category \
HAVING TotalCost > 150"
      
# Eksekusi perintah  SQL dengan Python (panda)
pd.read_sql(sql,db_enco)
```

## Soal 5

Kelompokkan inventory_items berdasarkan nama dan department produk dengan kasus total biaya untuk department laki-laki kurang dari 200 dan department perempuan lebih dari 200. Lalu urutkan dari yang terbesar

```python
# Perintah SQL 
sql = "SELECT product_name, product_department, SUM(cost) TotalCost \
FROM `inventory_items` \
GROUP BY product_name, product_department \
HAVING(SUM(CASE WHEN product_department = 'Men' \
           THEN cost ELSE NULL END) < 200 \
       OR SUM(CASE WHEN product_department = 'Women' \
              THEN cost ELSE NULL END) > 200) \
ORDER BY TotalCost DESC;"
    
# Eksekusi perintah  SQL dengan Python (panda)
pd.read_sql(sql,db_enco)
```
