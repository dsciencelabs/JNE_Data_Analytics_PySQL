# Berapa Penjualan setiap barang yang terdapat pada penjualan detail

```python
sql= '''SELECT barang.nama_barang, SUM(jml_barang)
FROM penjualan_detail
             INNER JOIN barang
                        ON penjualan_detail.id_barang = barang.id_barang
GROUP BY barang.id_barang'''

# Eksekusi perintah  SQL
pd.read_sql(sql, en_co)
```

#  Siapa Nama Pelanggan yang total transaksi paling banyak

```python
sql= '''SELECT pelanggan.nama, SUM(total_transaksi)
FROM penjualan 
             INNER JOIN pelanggan 
                        ON penjualan.id_pelanggan = pelanggan.id_pelanggan
GROUP BY pelanggan.id_pelanggan
ORDER BY SUM(total_transaksi) DESC'''

# Eksekusi perintah  SQL
pd.read_sql(sql, en_co)
```

# Nama Barang yang dibeli oleh setiap pelanggan dan jumlahnya

```python
sql= '''SELECT pl.nama, 
               barang.nama_barang,
               jml_barang
FROM pelanggan AS pl
INNER JOIN penjualan AS pn ON pl.id_pelanggan = pn.id_pelanggan
INNER JOIN penjualan_detail AS pd ON pn.id_transaksi= pd.id_transaksi
INNER JOIN barang ON pd.id_barang = barang.id_barang'''

# Eksekusi perintah  SQL
pd.read_sql(sql, en_co)
```

# Mencari dan Menunjukkan Status dari orders dengan berdasarkan nilai maximum dari order_id dengan menampilkan order_id, user_id, num_of_item, status. 

```python
import pandas as pd

sql = '''SELECT order_id, user_id, num_of_item, status FROM `orders` WHERE order_id = ( SELECT MAX(order_id) FROM `orders`);'''
pd.read_sql(sql, en_co)
```

# Mencari dan Menunjukkan nama product dan brand jika dengan berdasarkan retail price lebih besar dari rata-rata retail price keseluhurannya dalam data products dengan menampilkan id,name,brand

```python
sql = 'SELECT id,name,brand FROM `products` WHERE retail_price >(SELECT AVG(retail_price) FROM `products`);'
pd.read_sql(sql, en_co)
```

# Mencari dan Menunjukkan nama product jika dengan berdasarkan retail price lebih besar dari rata-rata retail price keseluhurannya dalam data products Serta berdasarkan brand "TYR" dengan menampilkan  id,cost,name,sku
