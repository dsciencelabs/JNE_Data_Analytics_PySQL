```python
# import module 
import mysql.connector

# koneksi salah satu database MySQl ke Python
db1 = mysql.connector.connect(
  host='localhost',
  user='bakti',
  passwd='123',
  database='jne2'
)

# membuat objek 'cursor` sebagai kata kunci eksekusi SQL
cursor1 = db1.cursor()
```

```python
# Perintah SQL 
sql = "SELECT SUM(cost) \
       FROM  inventory_items\
       WHERE product_department='Men'";
       
# Eksekusi perintah  SQL
cursor1.execute(sql)

# menampilkan daftar tabel
for data in cursor1:
  print(data) 
```

```python
# Perintah SQL 
sql = "SELECT DISTINCT ABS(cost) \
       FROM  inventory_items"
       
# Eksekusi perintah  SQL
cursor1.execute(sql)

# menampilkan daftar tabel
for data in cursor1:
  print(data) 
```

## Perhitungan Agregat

```python
# Perintah SQL 
sql = "SELECT COUNT(product_category) \
       FROM inventory_items \
       WHERE cost<5"
       
# Eksekusi perintah  SQL
cursor1.execute(sql)

# menampilkan daftar tabel
for data in cursor1:
  print(data) 
```

```python
# Perintah SQL 
sql = "SELECT product_department,  \
          AVG(cost) \
        FROM inventory_items \
        GROUP BY product_department;"
       
# Eksekusi perintah  SQL
cursor1.execute(sql)

# menampilkan daftar tabel
for data in cursor1:
  print(data) 
```

```python
# Perintah SQL 
sql = "SELECT product_category,  \
          AVG(cost) \
        FROM inventory_items \
        GROUP BY product_category;"
       
# Eksekusi perintah  SQL
cursor1.execute(sql)

# menampilkan daftar tabel
for data in cursor1:
  print(data) 
```

```python

# 5. Kelompokkan inventory_items berdasarkan kategori dan department produk dengan kasus total biaya 
# untuk department laki-laki lebih dari 150 dan department perempuan lebih dari 180 

# Perintah SQL 
sql = "SELECT product_category, product_department, SUM(cost) AS TotalCost \
FROM `inventory_items` \
GROUP BY product_category, product_department \
HAVING(SUM(CASE WHEN product_department = 'Men' \
           THEN cost ELSE NULL END) > 150 \
       OR SUM(CASE WHEN product_department = 'Women' \
              THEN cost ELSE NULL END) > 180);"


# Eksekusi perintah  SQL
cursor1.execute(sql)

# menampilkan daftar tabel
for data in cursor1:
  print(data) 
```
